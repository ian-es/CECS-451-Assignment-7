```python
"""
Ian Escalante
11/30/2023
CECS 451 - 02
Assignment 7

Implement a program to perform filtering in the hidden Markov model (HMM).

a. Assume that the hidden state variable and the evidence variable are binary variables.

b. The program should compute P(X_t|e_1:t) when e_1:t is given.

c. The program receives user input as a string containing independent variables in Figure 1 in that order.
    For example, ğ‘, ğ‘, ğ‘, ğ‘‘, ğ‘“, ğ‘’1, ğ‘’2 , ... , ğ‘’ğ‘¡
        Enter the values: 0.5,0.7,0.3,0.9,0.2,t,t
    means ğ‘ = 0.5, ğ‘ = 0.7, ğ‘ = 0.3, ğ‘‘ = 0.9, ğ‘“ = 0.2, ğ‘’1 = ğ‘¡, ğ‘’2 = ğ‘¡
    
d. Show the probability for these inputs:
    i.  0.5,0.7,0.3,0.9,0.2,t,t
    ii. 0.5,0.7,0.3,0.9,0.2,t,t,f
    
e. Please follow the output format. (Fix precisions using "{:.4f}".format)
    P(X_t|e_1:t) = <#.####,#.####>
"""

P_0 = (0,0)
transition = (0,0)
sensor = (0,0)
evidence = []

def main():
    get_inputs()
    print("P(X_t|e_1:t) = <{P_X_e1_t[0]:.4f},{P_X_e1_t[1]:.4}>".format(P_X_e1_t = filtering()))

def get_inputs():
    valid = False
    while not valid:
        input_str = input("Enter the values: ")
        valid = parse_inputs(input_str)
        if not valid:
            print("Please enter the values in the correct format.")
        
def parse_inputs(inputs):
    table = [0,0,0,0,0,0]
    evidence.clear()
    global P_0
    global transition
    global sensor 
    clean_inputs = inputs.split(",")

    if len(clean_inputs) < 6:
        return False
    
    for i in range(5):
        try:
            table[i] = float(clean_inputs[i])
        except ValueError:
            print("Value given is not a float.")
            return False
        if table[i] < 0 or table[i] >1:
            print("Improper value given for tables.")
            return False
            
    P_0 = (table[0],1-table[0])
    transition = (table[1],table[2])
    sensor = (table[3],table[4])
    
    for j in range(5,len(clean_inputs)):
        if clean_inputs[j] != 't' and clean_inputs[j] != 'f':
            print("Improper value given for evidence.")
            return False
        evidence.append(clean_inputs[j])
    return True
    
def filtering():
    P_x_e = P_0
    for e in evidence:
        P_x_e_prev = P_x_e
        
        # P(X_t|e_1:t) = alpha*P(e_t|X_t)*Sum_x_t(P(X_t|x_t-1)*P(x_t-1|e_1:t-1))
        # equations separated by vector x/y
        P_x_e = (sensor[0]*(transition[0]*P_x_e_prev[0] + transition[1]*P_x_e_prev[1]), 
                 sensor[1]*((1-transition[0])*P_x_e_prev[0] + (1-transition[1])*P_x_e_prev[1]))
        P_x_e = alpha(P_x_e)
        
    return P_x_e

def alpha(P_x_e):
    a = 1/(P_x_e[0] + P_x_e[1])
    P_x_e = (a*P_x_e[0],a*P_x_e[1])
    return P_x_e
    
main()


```
